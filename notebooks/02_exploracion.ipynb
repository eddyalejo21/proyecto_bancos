Iniciamos el proceso de limpieza para obtener el periodo de todos los archivos

```python
import re
import unicodedata
from datetime import datetime
```

```python
# Función para normalizar texto
def normalizar_texto(texto):
    """
    Convierte texto a mayúsculas, elimina tildes,
    espacios dobles y caracteres innecesarios.
    """
    if pd.isna(texto):
        return ""
    
    texto = str(texto).strip().upper()
    
    # Eliminar tildes
    texto = unicodedata.normalize("NFKD", texto)
    texto = "".join(c for c in texto if not unicodedata.combining(c))
    
    # Normalizar espacios
    texto = re.sub(r"\s+", " ", texto)
    
    return texto.strip()


MESES = {
    "ENE": "01", "ENERO": "01", "JAN": "01", "JANUARY": "01",
    "FEB": "02", "FEBRERO": "02", "FEBRUARY": "02",
    "MAR": "03", "MARZO": "03", "MARCH": "03",
    "ABR": "04", "ABRIL": "04", "APR": "04", "APRIL": "04",
    "MAY": "05", "MAYO": "05",
    "JUN": "06", "JUNIO": "06", "JUNE": "06",
    "JUL": "07", "JULIO": "07", "JULY": "07",
    "AGO": "08", "AGOSTO": "08", "AUG": "08", "AUGUST": "08",
    "SEP": "09", "SEPT": "09", "SEPTIEMBRE": "09", "SEPTEMBER": "09",
    "OCT": "10", "OCTUBRE": "10", "OCTOBER": "10",
    "NOV": "11", "NOVIEMBRE": "11", "NOVEMBER": "11",
    "DIC": "12", "DICIEMBRE": "12", "DEC": "12", "DECEMBER": "12",
}
```

```python
# Función para extraer periodo desde texto
def extraer_periodo_desde_texto(texto):
    """
    Extrae una fecha desde un texto y devuelve periodo en formato YYYY-MM.
    Soporta formatos como:
    - 28-FEB-2015
    - 28/02/2015
    - 2015-02-28
    - 31 DE MARZO DE 2026
    """
    
    texto_norm = normalizar_texto(texto)
    
    # Caso 1: formato YYYY-MM-DD o YYYY/MM/DD
    patron_yyyy_mm_dd = r"(\d{4})[-/](\d{1,2})[-/](\d{1,2})"
    match = re.search(patron_yyyy_mm_dd, texto_norm)
    
    if match:
        anio = match.group(1)
        mes = match.group(2).zfill(2)
        return f"{anio}-{mes}"
    
    # Caso 2: formato DD/MM/YYYY o DD-MM-YYYY
    patron_dd_mm_yyyy = r"(\d{1,2})[-/](\d{1,2})[-/](\d{4})"
    match = re.search(patron_dd_mm_yyyy, texto_norm)
    
    if match:
        mes = match.group(2).zfill(2)
        anio = match.group(3)
        return f"{anio}-{mes}"
    
    # Caso 3: formato DD-MES-YYYY o DD MES YYYY
    patron_dd_mes_yyyy = r"(\d{1,2})[-\s]+([A-Z]+)[-\s]+(\d{4})"
    match = re.search(patron_dd_mes_yyyy, texto_norm)
    
    if match:
        mes_texto = match.group(2)
        anio = match.group(3)
        
        if mes_texto in MESES:
            mes = MESES[mes_texto]
            return f"{anio}-{mes}"
    
    # Caso 4: formato DD DE MES DE YYYY
    patron_dd_de_mes_de_yyyy = r"(\d{1,2})\s+DE\s+([A-Z]+)\s+DE\s+(\d{4})"
    match = re.search(patron_dd_de_mes_de_yyyy, texto_norm)
    
    if match:
        mes_texto = match.group(2)
        anio = match.group(3)
        
        if mes_texto in MESES:
            mes = MESES[mes_texto]
            return f"{anio}-{mes}"
    
    return None
```

```python
# Función para extraer periodo desde un archivo Excel
def extraer_periodo_archivo(archivo, hojas_busqueda=None, max_filas=20, max_columnas=10):
    """
    Busca la fecha del boletín en las primeras filas y columnas del archivo.
    Primero intenta en BALANCE e INDICADORES.
    Si no encuentra, intenta extraer desde el nombre del archivo.
    """
    
    if hojas_busqueda is None:
        hojas_busqueda = ["BALANCE", "INDICADORES"]
    
    excel = pd.ExcelFile(archivo)
    
    for hoja in hojas_busqueda:
        if hoja not in excel.sheet_names:
            continue
        
        df_header = pd.read_excel(
            archivo,
            sheet_name=hoja,
            header=None,
            nrows=max_filas,
            usecols=range(max_columnas)
        )
        
        textos = []
        
        for valor in df_header.values.flatten():
            if pd.notna(valor):
                textos.append(str(valor))
        
        texto_completo = " ".join(textos)
        periodo = extraer_periodo_desde_texto(texto_completo)
        
        if periodo:
            return periodo
    
    # Fallback: intentar desde el nombre del archivo
    periodo_nombre = extraer_periodo_desde_texto(archivo.name)
    
    if periodo_nombre:
        return periodo_nombre
    
    return None
```

```python
# Definir categorías agregadas que no corresponden a bancos individuales
CATEGORIAS_EXCLUIR = [
    "BANCOS PRIVADOS GRANDES",
    "BANCOS PRIVADOS MEDIANOS",
    "BANCOS PRIVADOS PEQUENOS",
    "BANCOS PRIVADOS PEQUEÑOS",
    "TOTAL BANCOS PRIVADOS",
    "BANCOS PRIVADOS COMERCIALES",
    "BANCOS PRIVADOS CONSUMO",
    "BANCOS PRIVADOS VIVIENDA",
    "BANCOS PRIVADOS MICROCREDITO",
    "BANCOS PRIVADOS MICROCRÉDITO",
    "BANCA MULTIPLE",
    "BANCA MÚLTIPLE",
    "BANCOS PRIVADOS DE MICROEMPRESA"
]


def es_categoria_agregada(nombre):
    """
    Retorna True si el nombre corresponde a una categoría agregada
    y no a un banco individual.
    """
    nombre_norm = normalizar_texto(nombre)
    
    for categoria in CATEGORIAS_EXCLUIR:
        categoria_norm = normalizar_texto(categoria)
        
        if categoria_norm in nombre_norm:
            return True
    
    return False
```

```python
# Lista de palabras clave que suelen aparecer en nombres de bancos
PALABRAS_CLAVE_BANCOS = [
    "PICHINCHA",
    "GUAYAQUIL",
    "PACIFICO",
    "PACÍFICO",
    "PRODUBANCO",
    "AUSTRO",
    "BOLIVARIANO",
    "CITIBANK",
    "DINERS",
    "RUMINAHUI",
    "RUMIÑAHUI",
    "INTERNACIONAL",
    "LOJA",
    "MACHALA",
    "SOLIDARIO",
    "PROCREDIT",
    "AMAZONAS",
    "COMERCIAL DE MANABI",
    "COMERCIAL DE MANABÍ",
    "LITORAL",
    "COOPNACIONAL",
    "CAPITAL",
    "DELBANK",
    "ATLANTIDA",
    "ATLÁNTIDA",
    "CODESARROLLO",
    "VISIONFUND",
    "BANCO",
    "BP"
]


def parece_banco(nombre):
    """
    Evalúa si un texto parece ser un banco individual.
    """
    nombre_norm = normalizar_texto(nombre)
    
    if not nombre_norm:
        return False
    
    # Excluir categorías agregadas
    if es_categoria_agregada(nombre_norm):
        return False
    
    # Excluir textos administrativos
    textos_no_bancos = [
        "CODIGO",
        "CÓDIGO",
        "CUENTA",
        "DESCRIPCION",
        "DESCRIPCIÓN",
        "FECHA",
        "SALDO",
        "TOTAL ACTIVO",
        "TOTAL PASIVO",
        "PATRIMONIO"
    ]
    
    for texto in textos_no_bancos:
        if normalizar_texto(texto) == nombre_norm:
            return False
    
    # Si contiene palabras típicas de bancos
    for palabra in PALABRAS_CLAVE_BANCOS:
        if normalizar_texto(palabra) in nombre_norm:
            return True
    
    return False


def detectar_fila_bancos(archivo, hoja, max_filas=20):
    """
    Detecta automáticamente la fila donde están los nombres de bancos.
    Retorna el índice de fila en base 0.
    """
    
    df = pd.read_excel(
        archivo,
        sheet_name=hoja,
        header=None,
        nrows=max_filas
    )
    
    mejor_fila = None
    mejor_conteo = 0
    
    for idx, fila in df.iterrows():
        conteo_bancos = 0
        
        for valor in fila:
            if parece_banco(valor):
                conteo_bancos += 1
        
        if conteo_bancos > mejor_conteo:
            mejor_conteo = conteo_bancos
            mejor_fila = idx
    
    if mejor_fila is None or mejor_conteo < 3:
        return None
    
    return mejor_fila
```

```python
# Función para extraer bancos por hoja
def extraer_bancos_hoja(archivo, hoja):
    """
    Extrae los bancos reales de una hoja determinada.
    Devuelve un DataFrame con:
    - archivo
    - hoja
    - fila_excel
    - columna_excel
    - banco_original
    """
    
    fila_bancos = detectar_fila_bancos(archivo, hoja)
    
    if fila_bancos is None:
        return pd.DataFrame()
    
    df = pd.read_excel(
        archivo,
        sheet_name=hoja,
        header=None
    )
    
    fila = df.iloc[fila_bancos]
    
    bancos = []
    
    for col_idx, valor in fila.items():
        if pd.isna(valor):
            continue
        
        nombre = str(valor).strip()
        
        if parece_banco(nombre):
            bancos.append({
                "archivo": archivo.name,
                "hoja": hoja,
                "fila_excel": fila_bancos + 1,
                "columna_excel_num": col_idx + 1,
                "banco_original": nombre,
                "banco_normalizado_preliminar": normalizar_texto(nombre)
            })
    
    return pd.DataFrame(bancos)
```

INICIAMOS CON LA LIMPIEZA DE LOS DATOS NUMÉRICOS

```python
INDICADORES_OBJETIVO = [
    {
        "indicador": "activos_total",
        "hoja": "BALANCE",
        "patrones_busqueda": [
            "TOTAL ACTIVO"
        ],
        "unidad": "miles_usd"
    },
    {
        "indicador": "morosidad_total",
        "hoja": "INDICADORES",
        "patrones_busqueda": [
            "MOROSIDAD DE LA CARTERA TOTAL"
        ],
        "unidad": "porcentaje_decimal"
    },
    {
        "indicador": "roe",
        "hoja": "INDICADORES",
        "patrones_busqueda": [
            "RESULTADOS DEL EJERCICIO / PATRIMONIO PROMEDIO"
        ],
        "unidad": "porcentaje_decimal"
    },
    {
        "indicador": "indice_capitalizacion_neto",
        "hoja": "INDICADORES",
        "patrones_busqueda": [
            "INDICE DE CAPITALIZACION NETO",
            "ÍNDICE DE CAPITALIZACIÓN NETO"
        ],
        "unidad": "porcentaje_decimal"
    },
    {
        "indicador": "suficiencia_patrimonial",
        "hoja": "INDICADORES",
        "patrones_busqueda": [
            "PATRIMONIO + RESULTADOS",
            "ACTIVOS INMOVILIZADOS NETOS",
            "ACTIVOS INMOVILIZADOS",
            "(3) (6)"
        ],
        "unidad": "porcentaje_decimal"
    }
]
```

```python
# Función para limpiar valores numéricos
def limpiar_valor_numerico(valor):
    """
    Convierte valores provenientes de Excel a número.
    Soporta valores numéricos directos y textos con comas, espacios o símbolos.
    """
    
    if pd.isna(valor):
        return np.nan
    
    if isinstance(valor, (int, float, np.integer, np.floating)):
        return float(valor)
    
    texto = str(valor).strip()
    
    if texto == "":
        return np.nan
    
    # Quitar espacios
    texto = texto.replace(" ", "")
    
    # Quitar separadores de miles comunes
    texto = texto.replace(",", "")
    
    # Quitar símbolos no numéricos habituales
    texto = texto.replace("%", "")
    
    try:
        return float(texto)
    except ValueError:
        return np.nan
```

```python
# Función para encontrar fila de un indicador
def encontrar_fila_indicador(archivo, hoja, patrones_busqueda):
    """
    Busca la fila donde se encuentra un indicador.
    Retorna el índice de fila base 0.
    
    Si patrones_busqueda tiene varios elementos, todos deben estar presentes
    en la misma fila normalizada.
    """
    
    df = pd.read_excel(
        archivo,
        sheet_name=hoja,
        header=None
    )
    
    patrones_norm = [normalizar_texto(p) for p in patrones_busqueda]
    
    for idx, fila in df.iterrows():
        texto_fila = " ".join(
            normalizar_texto(valor)
            for valor in fila
            if pd.notna(valor)
        )
        
        if all(patron in texto_fila for patron in patrones_norm):
            return idx
    
    return None
```

```python
# Mostrar filas donde se encuentran los indicadores objetivo
for archivo in archivos_piloto:
    print("=" * 100)
    print("Archivo:", archivo.name)
    
    for config in INDICADORES_OBJETIVO:
        fila = encontrar_fila_indicador(
            archivo=archivo,
            hoja=config["hoja"],
            patrones_busqueda=config["patrones_busqueda"]
        )
        
        print(
            config["indicador"],
            "| Hoja:",
            config["hoja"],
            "| Fila Excel:",
            None if fila is None else fila + 1
        )
```

```python
# Función para extraer indicador desde un archivo
def extraer_indicador_archivo(archivo, config_indicador):
    """
    Extrae un indicador específico para todos los bancos detectados
    en la hoja correspondiente.
    """
    
    hoja = config_indicador["hoja"]
    indicador = config_indicador["indicador"]
    unidad = config_indicador["unidad"]
    patrones_busqueda = config_indicador["patrones_busqueda"]
    
    periodo = extraer_periodo_archivo(archivo)
    
    fila_indicador = encontrar_fila_indicador(
        archivo=archivo,
        hoja=hoja,
        patrones_busqueda=patrones_busqueda
    )
    
    if fila_indicador is None:
        print(f"⚠️ No se encontró indicador {indicador} en {archivo.name} - hoja {hoja}")
        return pd.DataFrame()
    
    bancos_hoja = extraer_bancos_hoja(archivo, hoja)
    
    if bancos_hoja.empty:
        print(f"⚠️ No se encontraron bancos en {archivo.name} - hoja {hoja}")
        return pd.DataFrame()
    
    df = pd.read_excel(
        archivo,
        sheet_name=hoja,
        header=None
    )
    
    registros = []
    
    for _, banco in bancos_hoja.iterrows():
        col_idx = int(banco["columna_excel_num"]) - 1
        
        valor = df.iloc[fila_indicador, col_idx]
        valor_limpio = limpiar_valor_numerico(valor)
        
        registros.append({
            "periodo": periodo,
            "banco_original": banco["banco_original"],
            "banco_normalizado": banco["banco_normalizado_preliminar"],
            "indicador": indicador,
            "valor": valor_limpio,
            "unidad": unidad,
            "archivo_origen": archivo.name,
            "hoja_origen": hoja,
            "fila_origen_excel": fila_indicador + 1,
            "columna_origen_excel": banco["columna_excel_num"]
        })
    
    return pd.DataFrame(registros)
```

```python

# Función para procesar un archivo financiero completo
def procesar_archivo_financiero(archivo):
    """
    Procesa un archivo Excel financiero y extrae todos los indicadores objetivo.
    """
    
    dataframes = []
    
    for config in INDICADORES_OBJETIVO:
        df_indicador = extraer_indicador_archivo(
            archivo=archivo,
            config_indicador=config
        )
        
        if not df_indicador.empty:
            dataframes.append(df_indicador)
    
    if not dataframes:
        return pd.DataFrame()
    
    return pd.concat(dataframes, ignore_index=True)
```

```python
# Procesar archivos piloto y construir dataset largo
dataset_long_piloto = []

for archivo in archivos_piloto:
    print("Procesando:", archivo.name)
    df_archivo = procesar_archivo_financiero(archivo)
    dataset_long_piloto.append(df_archivo)

dataset_long_piloto = pd.concat(dataset_long_piloto, ignore_index=True)

dataset_long_piloto.head(50)
```
