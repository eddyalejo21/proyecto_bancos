```python
#Este apartado se encarga de cargar el archivo excel y mostrar las hojas que contiene
import pandas as pd
from pathlib import Path


archivo = Path("../data/raw/FINANCIERO MENSUAL BANCA PRIVADA 2026_03.xlsx")

xls = pd.ExcelFile(archivo)

xls.sheet_names
```

```python
#Recorremos cada hoja del archivo y mostramos las primeras filas para entender su estructura
for hoja in xls.sheet_names:
    df_temp = pd.read_excel(archivo, sheet_name=hoja, header=None)
    print("HOJA:", hoja)
    print("Filas:", df_temp.shape[0], "Columnas:", df_temp.shape[1])
    display(df_temp.head(15))
```

```python
#Elegimos la hoja que contiene la palabra "ACTIVO" para analizar su contenido
for hoja in xls.sheet_names:
    df = pd.read_excel(archivo, sheet_name=hoja, header=None)
    
    coincidencias = df.apply(
        lambda col: col.astype(str).str.contains("ACTIVO", case=False, na=False)
    )
    
    if coincidencias.any().any():
        print("Encontrado en hoja:", hoja)
```

```python
#Una vez identificada la hoja que contiene la palabra "ACTIVO", cargamos esa hoja para analizar su contenido
hoja = "BALANCE"

df = pd.read_excel(archivo, sheet_name=hoja, header=None)

pd.set_option("display.max_rows", 80)
pd.set_option("display.max_columns", 50)

df.head(80)
```

```python
# A partir de la exploración visual del archivo, identificamos que los nombres 
# de los bancos están en la fila 7 y los valores de los activos en la fila 9,
# comenzando desde la columna 3. Con esta información, podemos extraer los datos
# relevantes para crear una tabla de activos por banco.

fila_bancos = 7
fila_activos = 9
col_inicio_bancos = 3

bancos = df.iloc[fila_bancos, col_inicio_bancos:]
valores = df.iloc[fila_activos, col_inicio_bancos:]

tabla_activos = pd.DataFrame({
    "banco": bancos.values,
    "indicador": "Activos",
    "valor": valores.values
})

tabla_activos
```

```python
# Limpiamos la tabla eliminando filas con valores
# nulos en la columna "banco" y formateando los datos
tabla_activos = tabla_activos.dropna(subset=["banco"])

tabla_activos["banco"] = (
    tabla_activos["banco"]
    .astype(str)
    .str.strip()
    .str.upper()
)

tabla_activos["valor"] = (
    tabla_activos["valor"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace(" ", "", regex=False)
)

tabla_activos["valor"] = pd.to_numeric(tabla_activos["valor"], errors="coerce").round(2)

#Controla la visualización de los números en la tabla para que se muestren con formato de miles y dos decimales
pd.set_option("display.float_format", "{:,.2f}".format)

tabla_activos
```

```python
# A partir de la exploración visual del archivo, identificamos que la fecha del periodo 
# se encuentra en la cabecera del archivo, por lo que podemos crear una función para 
# extraer esa información y agregarla a nuestra tabla de activos por banco.
import re
import pandas as pd
from datetime import datetime

def normalizar_mes_espanol(fecha_str):
    meses = {
        "ENE": "JAN",
        "FEB": "FEB",
        "MAR": "MAR",
        "ABR": "APR",
        "MAY": "MAY",
        "JUN": "JUN",
        "JUL": "JUL",
        "AGO": "AUG",
        "SEP": "SEP",
        "OCT": "OCT",
        "NOV": "NOV",
        "DIC": "DEC"
    }

    fecha_str = fecha_str.upper().strip()

    for mes_es, mes_en in meses.items():
        fecha_str = fecha_str.replace(mes_es, mes_en)

    return fecha_str


def parsear_fecha(fecha_str):
    fecha_str = normalizar_mes_espanol(fecha_str)

    formatos = [
        "%d/%m/%Y",   # 28/02/2015
        "%Y-%m-%d",   # 2015-02-28
        "%d-%b-%Y",   # 28-FEB-2015
        "%d/%b/%Y",   # 28/FEB/2015
    ]

    for formato in formatos:
        try:
            return datetime.strptime(fecha_str, formato)
        except ValueError:
            continue

    return None


def obtener_periodo_desde_cabecera(archivo, sheet_name=0, filas_cabecera=15):
    df_header = pd.read_excel(
        archivo,
        sheet_name=sheet_name,
        header=None,
        nrows=filas_cabecera
    )

    patron_fecha = (
        r"\b\d{2}/\d{2}/\d{4}\b|"      # 28/02/2015
        r"\b\d{4}-\d{2}-\d{2}\b|"      # 2015-02-28
        r"\b\d{2}-[A-ZÁÉÍÓÚÑ]{3}-\d{4}\b|"  # 28-FEB-2015
        r"\b\d{2}/[A-ZÁÉÍÓÚÑ]{3}/\d{4}\b"   # 28/FEB/2015
    )

    for fila in range(df_header.shape[0]):
        for col in range(df_header.shape[1]):
            valor = str(df_header.iloc[fila, col]).strip().upper()

            match = re.search(patron_fecha, valor)

            if match:
                fecha = parsear_fecha(match.group())

                if fecha:
                    return fecha.strftime("%Y-%m")

    raise ValueError(f"No se encontró una fecha válida en la cabecera del archivo: {archivo}")

```

```python
periodo = obtener_periodo_desde_cabecera(archivo)

tabla_activos["periodo"] = periodo

tabla_activos
```
